```python
import numpy as np

df = ipl.copy()

# Ensure sort order for appearances
df = df.sort_values(['matchId','inning','over','ball'])

# first row where each batsman appears in an innings
first_appearance = (df
    .drop_duplicates(subset=['matchId','inning','batsman'])
    .loc[:, ['matchId','inning','batsman','over','ball']]
    .sort_values(['matchId','inning','over','ball'])
)

# cumcount gives 0,1,2... → add 1 for positions
first_appearance['batting_position'] = (first_appearance
                                        .groupby(['matchId','inning'])
                                        .cumcount() + 1)

# merge back
df = df.merge(first_appearance[['matchId','inning','batsman','batting_position']],
              on=['matchId','inning','batsman'], how='left')

# verify
print(df[['matchId','inning','over','ball','batsman','batting_position']].head(12))
```

```python
#phase
def map_phase(over):
    if 0 <= over <= 6:   return 'Powerplay'
    if 7 <= over <= 15:  return 'Middle'
    return 'Death'

df['phase'] = df['over'].astype(int).map(map_phase)
print(df['phase'].value_counts())
```

```python
df = df[df['over'] <= 20]

df.head()
```

```python
df['balls_faced'] = (
    df.groupby(['matchId','inning','batsman'])['ball'].transform('count')
)
df.head()
```

```python
df['run_scored'] = (
    df.groupby(['matchId','inning','batsman'])['batsman_runs'].transform('sum')
)
df.head()
```

```python
df['strike_rate'] = df['run_scored']/df['balls_faced']*100
df.head()
```

```python
df['is_fours'] = (df['batsman_runs'] == 4).astype(int)
df['is_six'] = (df['batsman_runs'] == 6).astype(int)

bound_count = (df.groupby(['matchId','inning','batsman']).agg(
    fours = ('is_fours','sum'),
    six = ('is_six','sum')
).reset_index()
)


df.head()
```

```python
df['is_dot'] = (df['batsman_runs'] == 0).astype(int)
df
```

```python
df['total_dots'] = df.groupby(['matchId','inning','batsman'])['is_dot'].transform('sum')
df['dot_balls'] =df.groupby(['matchId','inning','batsman'])['is_dot'].transform('sum')
df['dot_ball_pct'] = np.where(
    df['balls_faced'] > 0,
    (df['dot_balls'] / df['balls_faced']) * 100,
    0
)

df.head()
```

```python
boundary_stats = (
    df.groupby(['matchId', 'inning', 'batsman'])
      .agg(
          fours=('is_fours', 'sum'),
          sixes=('is_six', 'sum'),
          balls_faced=('ball', 'count')   
      )
      .reset_index()
)

boundary_stats['boundary_pct'] = np.where(
    boundary_stats['balls_faced'] > 0,
    ((boundary_stats['fours'] + boundary_stats['sixes']) / boundary_stats['balls_faced']) * 100,
    0
)
df = df.merge(boundary_stats[['matchId', 'inning', 'batsman', 'boundary_pct']],
              on=['matchId', 'inning', 'batsman'], how='left')
```

```python

r = player_match_df['runs_scored']

player_match_df['is_30']  = (r >= 30).astype(int)
player_match_df['is_50']  = (r >= 50).astype(int)
player_match_df['is_100'] = (r >= 100).astype(int)
```

```python
dup_winner_cols = [c for c in player_match_df.columns if c.startswith('winner_')]
print("Dropping:", dup_winner_cols)

player_match_df = player_match_df.drop(columns=dup_winner_cols, errors='ignore')
```

```python
career_summary = (
    player_match_df
      .groupby('batsman')
      .agg(
          total_matches=('matchId','nunique'),
          total_runs=('runs_scored','sum'),
          total_balls=('balls_faced','sum'),
          total_fours=('fours','sum'),
          total_sixes=('sixes','sum'),
          total_outs=('out_flag','sum'),
          total_30_plus=('is_30','sum')
      )
      .reset_index()
)
```

```python
career_summary['avg_runs'] = career_summary['total_runs'] / career_summary['total_matches']

career_summary['avg_strike_rate'] = (
    (career_summary['total_runs'] / career_summary['total_balls']) * 100
)

career_summary['boundary_freq'] = (
    (career_summary['total_fours'] + career_summary['total_sixes']) /
    career_summary['total_balls']
)

career_summary['consistency_30_plus'] = (
    career_summary['total_30_plus'] / career_summary['total_matches']
)

career_summary['dismissal_rate'] = (
    career_summary['total_outs'] / career_summary['total_matches']
)
```

```python
career_summary
```

```python
pos_stats = (
    player_match_df
    .groupby(['batsman','batting_position'])
    .agg(avg_runs_at_pos=('runs_scored','mean'))
    .reset_index()
)

best_pos = (
    pos_stats
    .sort_values(['batsman','avg_runs_at_pos'], ascending=[True, False])
    .groupby('batsman')
    .first()
    .reset_index()
    .rename(columns={'batting_position':'position_preference'})
)

career_summary = career_summary.merge(
    best_pos[['batsman','position_preference']],
    on='batsman',
    how='left'
)

def label_position(pos):
    if pos in [1, 2]:
        return 'Opener'
    elif pos == 3:
        return 'Anchor'
    elif pos in [4, 5]:
        return 'Middle-order'
    else:
        return 'Finisher'

career_summary['position_label'] = (
    career_summary['position_preference'].apply(label_position)
)
```
